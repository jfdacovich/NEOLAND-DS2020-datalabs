## Piedra, papel, tijeras

Vamos a jugar contra nuestro ordenador al famoso juego.

```python
# Importa la función choice del módulo random
import random
# Asignación lista de opciones: 
l_opciones = ["piedra","papel","tijeras"]
# Numero de partidas
partidas = 5
# Para ganar un jugador debe completar o ganar 2 partidas
para_ganar = 2
# Las siguientes variables acumularan las partidas ganadas de cada participante
pc_win = 0
jugador_win = 0
empate = 0
# Define una función que devuelva aleatoriamente una de las 3 opciones.
# Esto corresponderá a la jugada de la máquina. Totalmente aleatoria.
def cpu_random():
    pc_random = random.choice(l_opciones) 
    return (pc_random) 

# Define una función que pregunte tu elección: 'piedra', 'papel' o 'tijeras'
def opciones_a_escoger():
    opcion_a_escoger = input ("Escoge una de las siguientes opciones: piedra, papel o tijeras! ")
    while opcion_a_escoger not in l_opciones:
        opcion_a_escoger = input ("Recuerda escoger piedra, papel , o tijeras ")
    return opcion_a_escoger

# Define una función que resuelva un combate. 
# Devuelve 0 si hay empate, 1 si gana la máquina, 2 si gana el jugador humano 
def resultado(opcion_a_escoger, pc_random):
    global pc_win
    global jugador_win
    global empate
    if opcion_a_escoger == "tijeras" and pc_random == "tijeras":
        empate +=1
        return 0
    elif pc_random == "piedra" and opcion_a_escoger == "tijeras":
        pc_win +=1
        return 1
    elif pc_random == "papel" and opcion_a_escoger == "piedra":
        pc_win +=1
        return 1
    elif pc_random == "tijeras" and opcion_a_escoger == "papel":
        pc_win +=1
        return 1
    elif opcion_a_escoger == "piedra" and pc_random == "tijeras":
        jugador_win +=1
        return 2
    elif opcion_a_escoger == "papel" and pc_random == "piedra":
        jugador_win +=1
        return 2
    elif opcion_a_escoger == "tijeras" and pc_random == "papel":
        jugador_win +=1
        return 2
    else:
        empate +=1
        return 0
    
# Define una función que muestre la elección de cada jugador y el estado de la partida
# Esta función debe utilizarse cada vez que se actualicen los puntos acumulados
def estatus ():
    print(f"La maquina ha elegido {pc_random}")
    print(f"El jugador ha elegido {opcion_a_escoger}")
    print(f"La maquina lleva ganadas {pc_win} partidas")
    print(f"El jugador lleva ganadas {jugador_win} partidas")
    

while jugador_win < para_ganar and pc_win < para_ganar and empate + jugador_win + pc_win < partidas:
    pc_random = cpu_random()
    opcion_a_escoger = opciones_a_escoger()
    resultado(opcion_a_escoger, pc_random)
    estatus()

# Crea un bucle que itere mientras ningún jugador alcance el mínimo de victorias
# necesarias para ganar. Dentro del bucle resuelve la jugada de la
# máquina y pregunta la del jugador. Comparalas y actualiza el valor de las variables
# que acumulen las partidas ganadas de cada participante. 

    
# Anuncia por consola el ganador del juego en función de quién tiene más victorias 
# aculumadas
    if jugador_win > pc_win and jugador_win > empate:
        print("El jugador tiene mas victorias acumuladas")
    elif pc_win > jugador_win and pc_win > empate:
        print("Estas siendo vencido por una maquina")
        print("La maquina acumula mas victorias")
    elif empate> jugador_win and empate > pc_win:
        print("Van quedando empatados")
```
