````python
# Importamos las libreras.
import random

# Declaramos las variables

value = random.randint(1,6)
count = 0
guess = 0

# Creamos el iterador
while guess != value:
    count += 1 #incremento
    guess = input("Adivina un numero entre 1 y 6: ")
    if guess.isnumeric():
        guess = int(guess)
else:
    print("has adivinado en {} ocasiones".format(count))
    # asi sera de otra manera print (f"has adivinado en {count} ocaciones")

````
